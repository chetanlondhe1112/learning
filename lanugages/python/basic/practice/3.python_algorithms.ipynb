#### 1Ô∏è‚É£ What is Time Complexity?
Time complexity describes how the runtime of an algorithm grows as the input size (n) increases.

It helps in comparing different algorithms to understand which one is more efficient.

#### 2Ô∏è‚É£ Understanding Big-O Notation
Big-O notation represents the upper bound (worst case) of an algorithm‚Äôs growth rate.

    Big-O Notation	Name	Example
    O(1)	Constant Time	Hash Table Lookups (dict.get())
    O(log n)	Logarithmic Time	Binary Search
    O(n)	Linear Time	Iterating through an array
    O(n log n)	Linearithmic Time	Merge Sort, Quick Sort
    O(n¬≤)	Quadratic Time	Nested Loops (Bubble Sort)
    O(2‚Åø)	Exponential Time	Recursive Fibonacci
    O(n!)	Factorial Time	Traveling Salesman Problem

#### 3Ô∏è‚É£ Examples of Time Complexity in Python

**üîπ Example 1: O(1) (Constant Time)**
No matter how large n is, it always takes the same amount of time.

    def get_first_element(arr):
        return arr[0]  # Always takes the same time
    
**‚úÖ Why O(1)? It doesn't depend on the input size n.**

**üîπ Example 2: O(n) (Linear Time)**
As n grows, the number of operations grows linearly.

    def print_all_elements(arr):
        for item in arr:
            print(item)  # Loops `n` times
    
**‚úÖ Why O(n)? If n = 1000, it runs 1000 times.**

**üîπ Example 3: O(log n) (Logarithmic Time)**
The number of operations shrinks as n grows (Binary Search).

    def binary_search(arr, target):
        left, right = 0, len(arr) - 1
        while left <= right:
            mid = (left + right) // 2
            if arr[mid] == target:
                return mid
            elif arr[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
        return -1
    
**‚úÖ Why O(log n)? Each step divides the array in half.**

**üîπ Example 4: O(n¬≤) (Quadratic Time)**
Nested loops cause O(n¬≤) complexity.

    def print_pairs(arr):
        for i in range(len(arr)):
            for j in range(len(arr)):
                print(arr[i], arr[j])  # Runs `n * n` times

**‚úÖ Why O(n¬≤)? If n = 1000, it runs 1,000,000 times!**

#### 4Ô∏è‚É£ Worst, Best, and Average Case Analysis

    Best Case (Œ© - Omega Notation): Minimum time required
    Worst Case (O - Big O Notation): Maximum time required
    Average Case (Œò - Theta Notation): Expected time

For example, in Binary Search:

    Best Case (Œ©(1)) ‚Üí If the element is found in the first attempt.
    Worst Case (O(log n)) ‚Üí If we keep searching until one element remains.

**‚úÖ Key Takeaways**

    O(1) (Constant Time) is the fastest
    O(n) (Linear Time) is common
    O(log n) (Logarithmic Time) is faster than O(n)
    O(n¬≤) (Quadratic Time) is slow
    O(2‚Åø) (Exponential) should be avoided
